```python
from gensim.models import Word2Vec as wtv
```

```python
model = wtv.load('ass-train')
print(model)
```

```python
for i in range(100): print(model.wv.index2word[i])
```

```python
model.wv.doesnt_match('technology computer foot wire'.split())
```

```python
model.wv.most_similar(positive = 'what', topn= 100)
```
